# Car Sales EDA with SQL

## Questions answered:

- q1 What are the top 10 car manufacturers by average consumer rating?
- q2 Out of the top 10 car manufacturers by average consumer rating, what are the top 10 best selling models?
- q3 What exterior colors do customers prefer the most?
- q4 Which states have the highest number of sales?
- q5 What is the best selling car by state?

```python
# Setup environment & load data 
import pandas as pd
import pandasql as ps

cars = 'data/Online_Car_Sales_Marketplace.csv'
carsdf = pd.read_csv(cars)
pd.set_option('display.max_columns', None)
carsdf.head()
```

```python
# Sample query DataFrame
q0 = """
SELECT *
FROM carsdf
LIMIT 5;
"""
result0 = ps.sqldf(q0)
display(result0)
```

```python
# What are the top 10 car manufacturers by average consumer rating?
q1 = """
SELECT DISTINCT Make, 
       ROUND(AVG(ConsumerRating), 2) AS AvgConsumerRating
FROM carsdf
WHERE ConsumerReviews >= 30
GROUP BY Make
ORDER BY AvgConsumerRating DESC
LIMIT 10;
"""
result01 = ps.sqldf(q1)
display(result01)
```

```python
# Out of the top 10 car manufacturers by average consumer rating,
# what are the top 10 best selling models?
q2 = """
SELECT Make,
       Model,
       COUNT(VIN) AS NumberOfSales
FROM carsdf
WHERE Make IN (
    SELECT Make
    FROM carsdf
    WHERE ConsumerReviews >= 30
    GROUP BY Make
    ORDER BY AVG(ConsumerRating) DESC
    LIMIT 10
)
GROUP BY Make, Model
ORDER BY NumberOfSales DESC
LIMIT 10;
"""
result02 = ps.sqldf(q2)
display(result02)
```

```python
# What exterior colors do customers prefer the most?
q3 = """
SELECT DISTINCT ExteriorColor,
       COUNT(ExteriorColor) AS Count
FROM carsdf
GROUP BY ExteriorColor
ORDER BY Count DESC
LIMIT 10;
"""
result03 = ps.sqldf(q3)
display(result03)
```

```python
# Which states have the highest number of sales?
q4 = """
SELECT DISTINCT State,
       COUNT(State) AS NumberOfSales
FROM carsdf
GROUP BY State
ORDER BY NumberOfSales DESC
LIMIT 10;
"""
result04 = ps.sqldf(q4)
display(result04)
```

```python
# What is the best selling car by state?
q5 = """
WITH RankedCars AS (
    SELECT State,
           Make,
           Model,
           ROW_NUMBER() OVER(PARTITION BY State ORDER BY COUNT(*) DESC) AS Rank
    FROM carsdf
    WHERE State NOT IN ('AZ-101',
                        'Bldg',
                        'Glens',
                        'Michigan',
                        'SE',
                        'Suite',
                        'US-12',
                        'US-169')
    GROUP BY State,
             Make,
             Model
)

SELECT State,
       Make,
       Model
FROM RankedCars
WHERE Rank = 1
       ;
"""
result05 = ps.sqldf(q5)
display(result05)
```
